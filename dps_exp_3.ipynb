```
# prompt: with and without anonymization

import pandas as pd

# Sample data (Indian names and information)
data = {
    'Name': ['Rafi shaik', 'Priyawarrior', 'Amit', 'Arjun', 'vikram', 'anashuyya'],
    'Age': [29, 35, 22, 26, 30, 21],  # Ages to be categorized
    'Gender': ['Male', 'Female', 'Male', 'Female', 'Male', 'Female'],
    'Blood Group': ['B+', 'O-', 'A+', 'AB+', 'B-', 'O+'],
    'City': ['Delhi', 'Mumbai', 'Bangalore', 'Kolkata', 'Chennai', 'Hyderabad'],
    'Pincode': ['110001', '400001', '560001', '700001', '600001', '500001'],
    'Email': ['rafi.shaik@email.com','priya.warrior@example.com', 'amit@example.com',
              'arjun.gupta@example.com', 'vikram.bhai@example.com', 'anashuyya@example.com'],
    'Salary': [45000, 55000, 60000, 40000, 70000, 65000],
    'Disease': ['None', 'Asthma', 'None', 'None', 'Hypertension', 'None']
}

# Creating DataFrame
df = pd.DataFrame(data)

# --- Without Anonymization ---
print("DataFrame without anonymization:")
print(df)


# --- With Anonymization ---
# Define bins for Age and corresponding labels
age_bins = [20, 30, 40, 50, 60]  # Age ranges
age_labels = ['20-30', '30-40', '40-50', '50+']  # Labels for the age ranges

# Use pandas.cut to categorize ages into these bins
df['Age'] = pd.cut(df['Age'], bins=age_bins, labels=age_labels, right=False)

# Mask the last digits of the pincode (replace last 2 digits with XX)
df['Pincode'] = df['Pincode'].apply(lambda x: x[:-2] + 'XX')

# Mask the email domain part (show only the username part)
def mask_email(email):
    username, domain = email.split('@')
    return username + '@*.com'

df['Email'] = df['Email'].apply(mask_email)


#Further Anonymization (Example: Replace names with generic identifiers)
# Create a list of generic names
generic_names = [f"Person {i}" for i in range(1, len(df) + 1)]

# Replace the names with generic names
df["Name"] = generic_names

# Replace sensitive information with generic values (Example for disease column)
df['Disease'] = df['Disease'].replace({'Asthma':'Condition1', 'Hypertension': 'Condition2'}, inplace=False)

print("\nDataFrame with anonymization:")
df
```
